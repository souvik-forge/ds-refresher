# 02 - NumPy Indexing, Slicing & Vectorization

The payoff of NumPy: expressive indexing + fast vectorized math.

```python
import numpy as np
rng = np.random.default_rng(0)
```

## 1. Basic indexing & slicing

```python
a = np.arange(24).reshape(4, 6)
print(a)
print('row 1:', a[1])
print('elt [2,3]:', a[2, 3])
print('cols 1-3:\n', a[:, 1:4])
print('every other row:\n', a[::2])
```

**Important:** slices are *views*, not copies. Modifying a slice modifies the original.

```python
v = a[0]
v[0] = 999
print(a)  # the original changed!

# Use .copy() to detach
c = a[0].copy()
c[0] = -1
print(a[0])
```

## 2. Boolean (mask) indexing

```python
x = rng.integers(0, 20, size=10)
print(x)
print('mask >10:', x > 10)
print('values:', x[x > 10])

# Modify in place
x[x > 10] = 0
print(x)
```

## 3. Fancy (integer) indexing

```python
a = np.arange(10) * 10
print(a)
idx = [1, 3, 5, 7]
print(a[idx])
```

## 4. Vectorized math (no Python loops!)

```python
x = np.arange(1_000_000)

# vectorized — fast
%timeit x * 2 + 1

# python loop — slow (don't do this)
%timeit [v*2 + 1 for v in x[:10000]]  # only 1% of data, still slower
```

## 5. Broadcasting
Different shapes can combine if compatible (trailing dims align, length-1 dims stretch).

```python
a = np.arange(12).reshape(3, 4)
print('a:\n', a)
print('a + 100:\n', a + 100)            # scalar broadcast
print('a + [0,1,2,3]:\n', a + np.array([0,1,2,3]))   # row broadcast
print('column add:\n', a + np.array([[0],[10],[20]]))  # column broadcast
```

## 6. Universal functions (ufuncs)

```python
x = np.linspace(0, np.pi, 5)
print(np.sin(x))
print(np.exp(x))
print(np.sqrt(np.arange(5)))
```
