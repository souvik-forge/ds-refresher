# 03 - NumPy Aggregations, Axes & Linear Algebra

```python
import numpy as np
rng = np.random.default_rng(7)
```

## 1. Aggregations: sum, mean, std, min/max
The key concept is the **axis**: which dimension to collapse.

```python
a = rng.integers(0, 10, size=(3, 4))
print(a)
print('sum all:', a.sum())
print('sum axis=0 (per column):', a.sum(axis=0))
print('sum axis=1 (per row):  ', a.sum(axis=1))
```

```python
print('mean:', a.mean())
print('std :', a.std())
print('min, argmin:', a.min(), a.argmin())
print('max per row:', a.max(axis=1))
```

## 2. Cumulative & sorting

```python
x = rng.integers(0, 10, size=8)
print(x)
print('cumsum :', np.cumsum(x))
print('sorted :', np.sort(x))
print('argsort:', np.argsort(x))   # indices that would sort
```

## 3. Stacking & splitting

```python
a = np.arange(6).reshape(2, 3)
b = np.arange(6, 12).reshape(2, 3)
print(np.vstack([a, b]))
print(np.hstack([a, b]))
print(np.concatenate([a, b], axis=0))
```

## 4. Linear algebra

```python
A = np.array([[1, 2], [3, 4]], dtype=float)
B = np.array([[5, 6], [7, 8]], dtype=float)

print('elementwise A*B:\n', A * B)
print('matmul A@B :\n', A @ B)
print('transpose  :\n', A.T)
print('inverse    :\n', np.linalg.inv(A))
print('determinant:', np.linalg.det(A))
```

```python
# Solve Ax = b
b = np.array([5, 11], dtype=float)
x = np.linalg.solve(A, b)
print('x =', x)
print('check A@x =', A @ x)
```

```python
# Eigen decomposition
eigvals, eigvecs = np.linalg.eig(A)
print('eigenvalues :', eigvals)
print('eigenvectors:\n', eigvecs)
```

## 5. Mini case study: standardize a dataset
Subtract the column mean, divide by the column std (a.k.a. z-score).

```python
data = rng.normal(loc=[10, 100, -5], scale=[2, 20, 1], size=(8, 3))
print('raw:\n', data.round(2))

z = (data - data.mean(axis=0)) / data.std(axis=0)
print('\nz-scored:\n', z.round(2))
print('means ~0:', z.mean(axis=0).round(6))
print('stds  ~1:', z.std(axis=0).round(6))
```
