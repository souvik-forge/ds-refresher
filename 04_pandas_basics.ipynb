# 04 - Pandas: Series, DataFrames, I/O & Selection

Pandas is the data-wrangling workhorse. Built on NumPy, it adds labeled rows/columns and missing-data handling.

```python
import numpy as np
import pandas as pd
print(pd.__version__)
```

## 1. Series — a 1D labeled array

```python
s = pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'], name='scores')
print(s)
print('s["b"] =', s['b'])
print('mean :', s.mean())
```

## 2. DataFrame — a 2D labeled table

```python
df = pd.DataFrame({
    'name'  : ['Ada', 'Bo', 'Cy', 'Di', 'Eve'],
    'age'   : [29, 35, 41, 22, 31],
    'city'  : ['NYC', 'SF', 'NYC', 'LA', 'SF'],
    'salary': [85_000, 120_000, 95_000, 60_000, 110_000],
})
df
```

```python
# Quick inspection
print(df.shape)
print(df.dtypes)
df.info()
df.describe()
```

## 3. Reading & writing files
Pandas reads almost anything: CSV, Excel, Parquet, JSON, SQL, HTML tables.

```python
# Write then read back as a demo
df.to_csv('/tmp/people.csv', index=False)
df2 = pd.read_csv('/tmp/people.csv')
df2
```

**Useful `read_csv` arguments:** `sep`, `header`, `names`, `usecols`, `dtype`, `parse_dates`, `na_values`, `nrows`, `chunksize`.

## 4. Selecting columns and rows
- `df['col']` → Series
- `df[['a','b']]` → DataFrame
- `df.loc[row_label, col_label]` → label-based
- `df.iloc[row_int,  col_int]`  → position-based

```python
print(df['name'])           # Series
print(df[['name', 'city']]) # DataFrame
```

```python
# Label-based
print(df.loc[0])                       # first row
print(df.loc[0:2, ['name', 'age']])    # inclusive on both ends!

# Position-based
print(df.iloc[0])
print(df.iloc[0:2, 0:2])               # exclusive on the right end
```

## 5. Boolean filtering

```python
df[df['age'] > 30]
```

```python
# Combine conditions — wrap each in parens, use & and |
df[(df['city'] == 'SF') & (df['salary'] > 100_000)]
```

```python
# .query() can read more naturally
df.query('city == "NYC" and age < 40')
```

## 6. Adding / transforming columns

```python
df['tax']         = df['salary'] * 0.30
df['salary_k']    = df['salary'] / 1000
df['senior']      = df['age'] >= 35
df
```
