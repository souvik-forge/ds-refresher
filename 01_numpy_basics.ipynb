## Why NumPy?
- Fast, vectorized numerical computation backed by C
- Homogeneous n-dimensional arrays (`ndarray`)
- Foundation for pandas, scikit-learn, PyTorch, etc.

```python
import numpy as np
print(np.__version__)
```

## 1. Creating arrays

```python
# From a Python list
a = np.array([1, 2, 3, 4])
print(a, a.dtype, a.shape)
```

```python
# 2D array
b = np.array([[1, 2, 3], [4, 5, 6]])
print(b)
print('shape:', b.shape, '| ndim:', b.ndim, '| size:', b.size)
```

```python
# Built-in constructors
print(np.zeros((2, 3)))
print(np.ones((2, 3), dtype=int))
print(np.full((2, 2), 7))
print(np.eye(3))
```

```python
# Ranges
print(np.arange(0, 10, 2))      # like range(), supports floats
print(np.linspace(0, 1, 5))     # 5 evenly spaced points incl. endpoints
```

## 2. dtypes
NumPy arrays are typed. Pick the right dtype to save memory and avoid surprises.

```python
x = np.array([1, 2, 3], dtype=np.float32)
print(x.dtype, x.nbytes, 'bytes total')

y = x.astype(np.int16)
print(y, y.dtype)
```

## 3. Random arrays
Use the modern `Generator` API (`np.random.default_rng`).

```python
rng = np.random.default_rng(seed=42)
print(rng.random((2, 3)))            # uniform [0,1)
print(rng.normal(0, 1, size=5))      # standard normal
print(rng.integers(0, 10, size=5))   # ints in [0,10)
```

## 4. Reshape & transpose

```python
a = np.arange(12)
print(a.reshape(3, 4))
print(a.reshape(3, 4).T)         # transpose
print(a.reshape(2, 2, 3).shape)  # 3D
```
